Q1.. Write a program which will find all such numbers which are divisible by 7 but are not mulltiple of 5 ,between 2000 and 3200(both include).The number obtained should be printed in acomma-seprated sequence of single line


```python

lists=[]
for i in range(2000, 3201):
    if (i%7==0) and (i%5!=0):
        lists.append(str(i))

print(','.join(lists))
lists=[]
```
